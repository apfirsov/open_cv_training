```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import easyocr
```

```python
model = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
img = cv2.imread('images/plate_1.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.equalizeHist(img)
plt.imshow(img)
```

```python
results = model.detectMultiScale(img, 1.05, 3)
print(len(results))
```

```python
plates = []
for (x, y, w, h) in results:
    #cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 1)
    plates.append(img[y:y+h, x:x+w])
```

```python
# for plate in plates:
#     plt.imshow(plate)
plt.imshow(plates[0])
```

```python
plt.imshow(plates[1])
```

```python
reader = easyocr.Reader(['ru', 'en'])
result = reader.readtext(plates[0])
print(result)
```

```python
result_2 = reader.readtext(plates[1])
print(result_2)
```

```python
result_img = reader.readtext(img)
print(result_img)
```
