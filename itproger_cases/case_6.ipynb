```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(image):
    converted_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    ax = plt.axes()
    ax.imshow(converted_img)
```

```python
photo = cv2.imread('images/img1.jpg')
img = np.zeros(photo.shape[:2], dtype='uint8')
circle = cv2.circle(img.copy(), (0, 0), 80, 255, -1)
square = cv2.rectangle(img.copy(), (25, 25), (550, 550), 255, -1)
```

```python
show(img)
```

```python
show(circle)
```

```python
show(square)
```

```python
img = cv2.bitwise_and(circle, square)
show(img)
```

```python
img = cv2.bitwise_or(circle, square)
show(img)
```

```python
img = cv2.bitwise_xor(circle, square)
show(img)
```

```python
img = cv2.bitwise_not(circle)
show(img)
```

```python
img = cv2.bitwise_not(square)
show(img)
```
